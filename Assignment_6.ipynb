```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
drive.mount('/content/drive')
import warnings
warnings.filterwarnings('ignore')
```

```python
gym= pd.read_csv('/content/drive/MyDrive/Data Science and ML/gym_membership.csv')
```

# **Bar Chart of Gym Membership Type**

```python
plt.figure()
counts = gym['abonoment_type'].value_counts()
plt.bar(counts.index, counts.values, color=['Red','Green'])
plt.xlabel("Membership Type")
plt.ylabel("Count")
plt.title("Gym Membership Types")
plt.show()
```

#  **Histogram of time in gym**

```python
plt.figure()
plt.hist(gym['avg_time_in_gym'],bins=40)
plt.xlabel("Average Time in Gym")
plt.ylabel("Frequency")
plt.title("Distribution of Time Spent in Gym")
plt.show()
```

# **Scatter plot: Age vs visits per week**

```python
plt.figure()
plt.scatter(gym['Age'], gym['visit_per_week'])
plt.grid(True)
plt.xlabel("Age")
plt.ylabel("Visits per Week")
plt.title("Age vs Visits per Week")
plt.show()
```

# **Box plot: Gender vs Age**

```python
sns.boxplot(data=gym, x='gender', y='Age')
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.7)
plt.title('Age by Gender')
plt.show()
```

# **Count plot: Days per week**

```python
sns.countplot(data=gym, x='days_per_week')
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.7)
plt.title('Days Per Week Count')
plt.show()
```

```python
plt.figure(figsize=(20, 10))
gym['days_per_week'].value_counts().sort_index().plot(kind='bar')
plt.xlabel("Days per Week")
plt.ylabel("Count")
plt.title("Gym Attendance Days per Week")
plt.show()
```

# **Bar plot: Average visits by membership type**

```python
sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.7)
plt.title('Average visits by membership type')
plt.show()
```
