# **Matplotlib**

## **Line Graph**

```python
# Simple line plot: First 10 math scores
plt.plot(students['math score'].head(10))  # X-axis auto (row numbers)

plt.title('Math Scores of First 10 Students')  # Title
plt.xlabel('Student Number')  # X label
plt.ylabel('Math Score')  # Y label
plt.show()  # Show plot
```

```python
# First 10 math scores
scores = students['math score'].head(10)

plt.plot( scores, marker='o')
plt.title('Math Scores of First 10 Students')
plt.xlabel('Student Number')
plt.ylabel('Math Score')
plt.savefig('/content/drive/MyDrive/Data Science and ML/Math_Score.jpg', dpi=300, bbox_inches="tight") # we can save the figure in our drive
plt.show()
```

## **Bar Chart**

```python
# Count gender
gender_count = students['gender'].value_counts()

# Simple bar chart
plt.bar(['Female', 'Male'], gender_count.values)
plt.title('Gender Count')
plt.ylabel('Count')
plt.xlabel('Gender')
plt.show()
```

```python
# With color
plt.bar(['Female', 'Male'], gender_count.values, color=['Red', 'green'])
plt.ylabel('Count')
plt.title('Students by Gender')
plt.show()
```

```python
# Group by gender, average math score
avg_math = students.groupby('gender')['math score'].mean()

# Bar plot
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.7)
plt.bar(avg_math.index, avg_math.values, color=['black', 'purple'])
plt.title('Average Math Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Math Score')
plt.show()
```

## **Scatter Plot**

```python
# Math vs Reading
plt.scatter(students['math score'], students['reading score'])
plt.grid(True)
plt.xlabel('Math')
plt.ylabel('Reading')
plt.title('Math vs Reading')
plt.show()
```

```python
# Smaller sample for clarity
sample = students.head(100)
plt.scatter(sample['math score'], sample['reading score'], alpha=0.5)
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.show()
```

## **Histogram**

```python
# Math score distribution
plt.hist(students['math score'])
plt.title('Math Score Distribution')
plt.show()
```

```python
# With more bins
plt.hist(students['math score'], bins=20, color='green')
plt.xlabel('Score')
plt.ylabel('Count')
plt.title('Math Scores')
plt.show()
```

## **Pie Chart**

```python
# Test preparation
test_prep = students['test preparation course'].value_counts()

plt.pie(test_prep.values, labels=['None', 'Completed'])
plt.title('Test Preparation')
plt.show()
```

```python
# With percentages
plt.pie(test_prep.values, labels=['None', 'Completed'], autopct='%1.1f%%')
plt.title('Test Prep Status')
plt.show()
```

```python
education_counts = students['parental level of education'].value_counts()
plt.pie(education_counts, labels=education_counts.index, autopct='%1.2f%%')
plt.title('Parental Education Levels')
plt.show()
```

## **Multiple Plots**

```python
# Create 2x2 grid
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plot 1: Histogram
axes[0, 0].hist(students['math score'], bins=15)
axes[0, 0].set_title('Math Score')

# Plot 2: Bar chart
students['gender'].value_counts().plot(kind='bar', ax=axes[0, 1])
axes[0, 1].set_title('Gender')

# Plot 3: Scatter
axes[1, 0].scatter(students['math score'][:50], students['reading score'][:50])
axes[1, 0].set_title('Math vs Reading')

# Plot 4: Line
axes[1, 1].plot(students['writing score'][:30])
axes[1, 1].set_title('Writing Scores')

plt.tight_layout()
plt.show()
```

# **Seaborn**

## **Histogram**

```python
sns.histplot(students['math score'])
plt.title('Math Score with Seaborn')
plt.show()

# With KDE curve, KDE= Kernel Density Estimate

sns.histplot(students['math score'], kde=True)
plt.show()
```

## **Box Plot**

```python
# Gender vs Math Score
sns.boxplot(data=students, x='gender', y='math score')
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.7)
plt.title('Math by Gender')
plt.show()

# Lunch type effect
sns.boxplot(data=students, x='lunch', y='math score')
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.7)
plt.title('Math by Lunch Type')
plt.show()
```
