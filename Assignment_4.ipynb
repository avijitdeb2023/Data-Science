```python
import pandas as pd
import numpy as np
data = np.random.randint(1, 101, size=(5, 3))
df = pd.DataFrame(data, columns=['A', 'B', 'C'])
print("DataFrame:\n",df)
column_sum = df.sum()
print("\nSum of each column:")
print(column_sum)
column_mean = df.mean()
print("\nMean of each column:")
print(column_mean)





```
