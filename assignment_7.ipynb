```python
from sklearn.preprocessing import LabelEncoder



# Initialize label encoder
le = LabelEncoder()

# Automatically detect all object (non-numeric) columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Apply Label Encoding to all categorical columns
for col in categorical_cols:
    df[col] = le.fit_transform(df[col].str.lower())

print("Categorical columns encoded automatically:")
print(df.head())

```

```python
# Separate input (X) and output (y)
X = df.drop('price', axis=1)  # Everything except price
y = df['price']
print("X has all House information (12 columns)")
print("y has Price (1 column)")
# Split: 80% for training, 20% for testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,    # 20% for test
    random_state=42   # Same result every time
)
print(X_train.shape, X_test.shape)
print(f"\nWe will train with: {len(X_train)} Houses")
print(f"We will test with: {len(X_test)} Prices")
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Create scaler
scaler = StandardScaler()

# Learn from training data and scale it
X_train_scaled = scaler.fit_transform(X_train)

# Scale test data (just transform, don't fit)
X_test_scaled = scaler.transform(X_test)


print("\nAfter scaling:")
print("All features now have similar range!")
print(f"Example - First House's Area before scaling: {X_train.values[0][0]}")
print(f"Example - First House's Area after scaling: {X_train_scaled[0][0]:.2f}")

```

```python
from sklearn.linear_model import LinearRegression

# Create model
model = LinearRegression()

# Train the model
model.fit(X_train_scaled, y_train)
print("✅ Model trained!")

# Evaluate
train_score = model.score(X_train_scaled, y_train)
print(f"Training accuracy: {train_score:.1%}")

test_score = model.score(X_test_scaled, y_test)
print(f"Test accuracy: {test_score:.1%}")


```
